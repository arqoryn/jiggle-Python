# DATA CLEANING

## Cleaning 'Item' column

```python
#item
df.dropna(subset="Item", inplace=True) # drop NA rows
for i in df.index:
  if df.loc[i, "Item"] == 'UNKNOWN' or df.loc[i, "Item"] == 'ERROR':
    df.drop(i, inplace= True) #drop UNKNOWN, ERROR rows

df['Item'] = df['Item'].astype('string')
print(df)
print(df.info())

```

## Cleaning 'Quantity' column

```python
#quantity
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")

quantity_mean = round(df["Quantity"].mean())
df.fillna({"Quantity":quantity_mean}, inplace=True)

print("Quantity mean: ", quantity_mean)

print(df)
print(df.info())
```

## Cleaning 'Price per unit' column

```python

#Price per unit

df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors="coerce")
price_mode = df["Price Per Unit"].mode()[0]
df.fillna({"Price Per Unit":price_mode}, inplace=True)

print("Price per Unit mode: ", price_mode)

print(df)
print(df.info())
```

## Cleaning 'Total Spent' column

```python
# total spent
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")

# Calculate 'Total Spent' for NaN values
# Total Spent is Quantity * Price Per Unit (common for sales data)
mask_nan_total_spent = df["Total Spent"].isna() # boolean Mask -> [False, False, False, True, False]
df.loc[mask_nan_total_spent, "Total Spent"] = df.loc[mask_nan_total_spent, "Quantity"] * df.loc[mask_nan_total_spent, "Price Per Unit"]

print(df)
print(df.info())
```

## Cleaning 'Payment Method' column

converting ERROR and UNKNOWN to NaN using Numpy

```python
# Payment Method
import numpy as np
df["Payment Method"] = df["Payment Method"].replace('UNKNOWN',np.nan)
df["Payment Method"] = df["Payment Method"].replace('ERROR',np.nan)
print(df)
```

lets see how many missing values there are in Payment Method, so that we can determine the solution


```python
missing_values = df["Payment Method"].isna().sum()
missing_perc = missing_values / len(df) * 100
print(missing_perc)
```

since the number is too high, we cant set it to mode, so we introdduce another type - Unknown

```python
df["Payment Method"] = df["Payment Method"].fillna("Unknown")    # Fill missing values with unknown
df["Payment Method"] = df["Payment Method"].astype('string')
print(df)
df.info()
```

## Cleaning 'Location' column

```python
# location
df["Location"] = df["Location"].replace('UNKNOWN', np.nan)
df["Location"] = df["Location"].replace('ERROR', np.nan)


df["Location"] = df["Location"].fillna('Unknown')
df["Location"] = df["Location"].astype('string')
print(df)
print(df.info())
```

## Cleaning 'Date' column

```python
# date
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], format='mixed', errors= 'coerce')
print(df)
print(df.info())
```

```python
# filling the NaT rows with mode date
mode_date = df["Transaction Date"].mode()[0]
df["Transaction Date"] = df["Transaction Date"].fillna(mode_date)
print(df)
df.info()
```
