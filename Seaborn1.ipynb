# Matrix Plots

## Heat Maps

### Corelation Method

```python
cordf = df.corr(numeric_only=True)
sns.heatmap(cordf, annot=True, cmap='Blues')
```

### Pivot Table Method

```python
pvdf = df.pivot_table(index='Parental_Involvement', columns='Parental_Education_Level', values='Exam_Score')

# Define the desired order for index and columns
parental_involvement_order = ['Low', 'Medium', 'High']
parental_education_order = ['High School', 'College', 'Postgraduate']

# Reindex the DataFrame
pvdf = pvdf.reindex(index=parental_involvement_order, columns=parental_education_order)


# in good ordered data YOU JUST DO THIS
sns.heatmap(pvdf, cmap='Blues', linecolor='white', linewidth=1)
```
