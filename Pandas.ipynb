**Calculate MEAN and then replace**

````python
df = pd.read_csv("datacleaning.csv")

x = df["Calories"].mean()
df.fillna({"Calories":x}, inplace= True)

print("Mean value of Calories = ", x)
print(df)
````

**Calculate MEDIAN and then replace**

````python
df = pd.read_csv("datacleaning.csv")

y = df["Calories"].median()
df.fillna({"Calories":y}, inplace= True)

print("Median of Calories: ", y)
print(df)
````

**Calculate MODE and then replace**

````python
df = pd.read_csv("datacleaning.csv")

z = df["Calories"].mode()[0]
df.fillna({"Calories":z}, inplace=True)

print("Mode of Calories: ", z)
print(df)
````

### Wrong Format

Cells with data of wrong format can make it difficult, or even impossible, to analyze data.

To fix it, you have two options: remove the rows, or convert all cells in the columns into the same format.

**Convert into same format** -> *to_datetime()*

````python
df = pd.read_csv("datacleaning.csv")

df["Date"] = pd.to_datetime(df["Date"], format='mixed')

print(df)
````

**Remove NaT rows**

````python
df.dropna(subset='Date', inplace = True)
print(df)
````

### Wrong Data

Option 1: Replacing the values

````python
# 'Duration' column mai '7th' row ka value = '45' hai
df.loc[7, "Duration"] = 45

print(df.loc[7])
print(df)
````

Auto-Replacing for large datasets (rule based - if duration > 120 -> set 120)


````python
df = pd.read_csv("datacleaning.csv")
for x in df.index:
  if df.loc[x, "Duration"] > 80:
    df.loc[x, "Duration"] = 80

print(df)
````

Option 2: Removing the row if duration > 120 (not likely)

````python
df = pd.read_csv("datacleaning.csv")

for i in df.index:
  if df.loc[i, "Duration"] > 120:
    df.drop(i, inplace=True)

print(df)
````

### Remove Duplicates

Discovering duplicated rows using *duplicated()*

````python
print(df.duplicated())
````

We now know row 12 is a duplicate

using *drop_duplicates()* to remove duplicate rows

````python
df.drop_duplicates(inplace=True)
print(df)
````

## Final Cleaning

````python
df = pd.read_csv("datacleaning.csv")

# clean Calories
x = df["Calories"].mean()
df.fillna({"Calories":x}, inplace=True)

# clean Date
df['Date'] = pd.to_datetime(df['Date'], format= 'mixed')
df.dropna(subset="Date", inplace= True)

# clean Duration
for i in df.index:
  if df.loc[i, "Duration"] > 120:
    df.loc[i, "Duration"]  = 120

# clean duplicates
df.drop_duplicates(inplace= True)

# print cleaned data
print(df)
````

```markdown
This comprehensive cleaning script:

1.  **Handles missing 'Calories' data** by imputing them with the mean value.
2.  **Converts 'Date' column to a proper datetime format** and removes rows with unparseable dates.
3.  **Corrects 'Duration' outliers** by capping values greater than 120 to 120.
4.  **Removes duplicate rows** to ensure uniqueness in the dataset.
```
